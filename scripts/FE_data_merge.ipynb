```python
import pandas as pd 
import numpy as np
import glob
import os
```

```python
def get_raw_data(directory):

    csv_files = glob.glob(os.path.join(directory, "*.csv"))
    xlsx_files = glob.glob(os.path.join(directory, "*.xlsx"))
    final_df = pd.DataFrame(columns=['model_year', 'manufacturer_name', 'vehicle_model', 'displacement', 'num_cylinders', \
                            'city_FE', 'hwy_FE', 'combine_FE', 'drive_system','fuel_type', 'vehicle_type'])
    for csv_item in csv_files:
        file_name = csv_item[32:] # Get the file name of each path
        if int(file_name[:4]) in [2002, 2003, 2004, 2005, 2007, 2008, 2009]:
            df = pd.read_csv(csv_item, usecols = [0,1,2,3,4,6,8,9,10,14])
            df = df.iloc[:, [1,2,3,4,6,7,8,5,9,0]]
            new_col = pd.Series([int(file_name[:4])]*len(df))
            df.insert(0, 'model_year', new_col)
            df.columns = final_df.columns.tolist()
            final_df = pd.concat([final_df, df], ignore_index=True)
        else:
            df = pd.read_csv(csv_item, usecols = [0,1,2,3,4,6,7,8,9,13])
            df = df.iloc[:, [1,2,3,4,6,7,8,5,9,0]]
            new_col = pd.Series([int(file_name[:4])]*len(df))
            df.insert(0, 'model_year', new_col)
            df.columns = final_df.columns.tolist()
            final_df = pd.concat([final_df, df], ignore_index=True)

    for xlsx_item in xlsx_files:
        file_name = xlsx_item[32:] # Get the file name of each path
        if int(file_name[:4]) in list(range(2013,2026)):
            df = pd.read_excel(xlsx_item, sheet_name=0, usecols = [0,2,3,6,7,9,10,11,27,33,69])
            df.columns = final_df.columns.tolist()
            final_df = pd.concat([final_df, df], ignore_index=True)
        else:
            df = pd.read_excel(xlsx_item, sheet_name=0, usecols = [0,2,3,6,7,9,10,11,27,33,66])
            df.columns = final_df.columns.tolist()
            final_df = pd.concat([final_df, df], ignore_index=True)
    
    return final_df
    
# Find all xlsx and CSV files in the directory
directory = "../fuel_economy_data/extract_to/"  
raw_data = get_raw_data(directory)
raw_data.head()

```

```python
raw_data.to_csv("../fuel_economy_data/merge_raw_data.csv", index=False)
```
